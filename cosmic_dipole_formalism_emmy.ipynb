# --- Step 1: Generate Mock Galaxy Catalog ---

````python
def proba_schechter_mag(M, M_star, alpha):
    """Compute the relative probability of a magnitude M.
    M_star is the caracteristic magnitude of Schechter's luminosity function."""
    L_ratio = 10**(0.4 * (M_star - M)) #Conversion magnitude -> relative luminosity L/L*
    return (L_ratio**alpha) * np.exp(-L_ratio)

def generate_schechter_magnitudes(N, M_min=-23, M_max=-15, M_star=-20.44, alpha=-1.1):
    """Generate N randomized absolute magnitudes, by using a reject test and Schechter's law."""
    magnitudes = []
    while len(magnitudes) < N:
        M_test = np.random.uniform(M_min, M_max)
        p_test = np.random.uniform(0, 1) #Probability for rejet test
        if p_test < proba_schechter_mag(M_test, M_star, alpha): magnitudes.append(M_test)         
    return np.array(magnitudes)

def get_dL(zi, H0=67.4, Om=0.315, Ol=0.685, c=3e8):
    '''Compute the luminosity distance depending on the redshift z, the Hubble constant H0, the cosmological parameters Om and Ol, and the speed of ligt c.'''
    inv_Ez = lambda zp: 1.0 / np.sqrt(Om * (1 + zp)**3 + Ol)
    integral, _ = quad(inv_Ez, 0, zi)
    return (c / H0) * (1 + zi) * integral
````

````python
def generate_galaxies(N, z_min=0.01, z_max=2.0, M_min=-23, M_max=-15, M_star=-20.44, alpha=-1.1, sigma_M_star=1.5, rejectTest_M=False, only_coord=False, **kwargs):
    #Cosmological Parameters  (Planck 2018):
    H0 = kwargs.get('H0', 67.4)
    Om = kwargs.get('Om', 0.315)
    Ol = kwargs.get('Ol', 0.685)
    c = kwargs.get('c', 3e5) #speed of ligt in km/s (because H0 is in Km/s/Mpc)
    M_sun = kwargs.get('M_sun', 4.83) #Absolute magnitude of Sun

    #Coord ra, dec, z:
    ra = np.random.uniform(0, 360, N)
    dec = np.random.uniform(-1, 1, N)
    dec = np.degrees(np.arcsin(dec))
    z = np.random.uniform(z_min, z_max, N) #Redshift
    if only_coord: return ra, dec, z

    #Other data:
    dL_mpc = np.array([get_dL(zi, H0, Om, Ol, c) for zi in z]) #luminosity distance in Mpc
    #Absolute magnitudes:
    if rejectTest_M: M = generate_schechter_magnitudes(N, M_min, M_max, M_star, alpha)
    else: M = np.random.normal(M_star, sigma_M_star, N)
    m = M + 5*np.log10(dL_mpc) + 25 #Aparent magnitudes
    L = 10**(0.4 * (M_sun - M)) #luminosities
    return ra, dec, z, m, M, L, dL_mpc
````

# --- Step 2: Apply Kinematic Dipole ---

````python
def apply_dipole(ra, dec, z, v=370*u.km/u.s, dipole_dir=(264*u.deg, 48*u.deg), c=3e5*u.km/u.s):
    """Returned the measured RA, DEC coorrdinates of sources, when they are modified by a kinematic dipole induced by the speed v of the observer, depending on the true ra, dec coordinates of the sources."""
    sources = SkyCoord(ra=ra*u.deg, dec=dec*u.deg, frame='icrs')
    u_source = sources.cartesian.xyz.value #unitary ectors x, y, z for each source: shape = (3, N)
    dipole = SkyCoord(l=dipole_dir[0], b=dipole_dir[1], frame='galactic').icrs
    u_dipole = dipole.cartesian.xyz.value
    #print("u_dipole = ", hp.ang2vec(dipole.ra.deg, dipole.dec.deg, lonlat=True))
    #print("u_dipole = ", u_dipole)
    
    beta_vec = (v/c).to(u.dimensionless_unscaled).value * u_dipole
    beta = np.linalg.norm(beta_vec)
    
    costheta = np.dot(u_dipole, u_source) #theta = angle between the "true" direction and the observed direction.
    factor = (1 + costheta * beta) # Doppler factor for beta << 1.
    u_new = (u_source + beta_vec[:, np.newaxis] if u_source.ndim > 1 else u_source + beta_vec) / factor # New vectors for each source: shape = (3, N)

    new_c = SkyCoord(x=u_new[0], y=u_new[1], z=u_new[2], representation_type='cartesian', frame='icrs')
    #print('norm u_new =', np.linalg.norm(u_new, axis=0))
    #return np.new_c.spherical.lon.deg, new_c.spherical.lat.deg
    return hp.vec2ang(u_new.T, lonlat=True)

def apply_dipole_map(map, M, D, nest, radec=False):
    """Returned the measured map, when it is modified by a a monopole M and a kinematic dipole D, depending on the true map."""
    D = np.array(D)
    npix = len(map) #nb. of pixels
    nside= hp.npix2nside(npix)
    ipix = np.arange(npix)
    if radec:
        thetaD, phiD = np.radians(90-D[2]), np.radians(D[1])
        D = D[0]*np.array([np.sin(thetaD)*np.cos(phiD), np.sin(thetaD)*np.sin(phiD), np.cos(thetaD)])
        ra, dec = hp.pix2ang(nside, ipix, nest=nest, lonlat=True)
        theta, phi = np.radians(90-dec), np.radians(ra) 
        u_source = np.array([np.sin(theta)*np.cos(phi), np.sin(theta)*np.sin(phi), np.cos(theta)])
    else: u_source = hp.pix2vec(nside, ipix, nest)
    #u_dipole = D/np.linalg.norm(D)
    #print("u_dipole = ", hp.ang2vec(dipole.ra.deg, dipole.dec.deg, lonlat=True))
    #print("u_dipole = ", u_dipole)
    #print('u_source =', u_source)
    return M + np.dot(D, u_source)
````

````python
def model_dipole_lb(ipix, A, l, b, **kwargs):
    #print("ipix =", ipix)
    npix = len(ipix) #nb. of pixels
    nside= hp.npix2nside(npix)
    u_source = np.array(hp.pix2vec(nside, ipix.astype(int), **kwargs)) #position verctor for each pixel ; not unitary
    dec = (b % 180) - 90 #to be sure it stays in the correct interval
    u_dipole = hp.ang2vec(l, dec, lonlat=True) #dipole unitary vector
    costheta = np.sum(u_source.T * u_dipole, axis=1) #the sum is on x, y, z component (dot prodct): shape = npix
    #print('norm u_source =', np.linalg.norm(u_source, axis=0))
    return (1 + A * costheta)

def model_dipole_vlb(nside, ipix, v, l, b, z, c=3e5, **kwargs): #c in km/s
    #print("ipix =", ipix)
    #npix = len(ipix) #nb. of pixels
    #print("npix =",npix)
    #nside= hp.npix2nside(npix)
    ra, dec = hp.pix2ang(nside, ipix.astype(int), lonlat=True, **kwargs)
    ra_new, dec_new = apply_dipole(ra, dec, z, v*u.km/u.s, dipole_dir=(l*u.deg, b*u.deg), c=c*u.km/u.s)
    return RADEC2NSource(nside, ra_new, dec_new)
    #return hp.ang2pix(nside, ra_new, dec_new, lonlat=True)

def model_dipole_MD(map, M, Dx, Dy, Dz, nest, radec=False):
    return apply_dipole_map(map, M, [Dx, Dy, Dz], nest, radec)

def fit_dipole_err(model, map, init, names, bounds=([0, 0, -90], [1, 360, 90]), fixed=[], errSyst=0, delta_x=1, **kwargs):
    map_errY = np.sqrt(map)
        
    npix = len(map) #nb. of pixels
    ipix = np.arange(npix) #indices of each pixel
    #print("ipix =", ipix)
    nside= hp.npix2nside(npix)

    m = fit_minuit(ipix, map, map_errY, model, init, names, list(zip(bounds[0], bounds[1])), fixed, title="Fit dipole", xlabel="Pixels")
    popt, pcov = curve_fit(model, ipix, map, p0=init, bounds=bounds, sigma= map_errY)
    return popt, pcov, m
````

# --- Step 3: Apply Milky Way Extinction with Spatially Varying R_V ---

````python
def apply_extinction(ra, dec, m, ebv_map=None, rv_map=None):
    # Placeholder: Simulate E(B-V) and R_V maps
    if ebv_map is None:
        ebv = np.random.normal(0.03, 0.01, len(ra))
    if rv_map is None:
        rv = np.random.normal(3.1, 0.3, len(ra))
    A_r = rv * ebv * 2.753
    m_corrected = m + A_r
    return m_corrected
````

# --- Step 4: Apply LSST Selection Function ---

````python
def apply_lsst_selection(ra, dec, m, z):
    in_footprint = (dec > -60) & (dec < 30)
    detected = m < 27.5
    return in_footprint & detected
````

# --- Step 5: Tomographic Dipole Fitting ---

````python
def fit_dipole_tomographic(ra, dec, z, weights, z_bins=[0.1, 0.5, 1.0, 2.0], nest=True):
    results = []
    nside = 32
    npix = hp.nside2npix(nside)
    
    def model(pix, A, l, b):
        pix.dtype = int #because hp.pix2vec needs pix as int
        dipole_dir = hp.ang2vec(np.radians(90-b), np.radians(l))
        pix_vecs = np.array(hp.pix2vec(nside, pix))
        costheta = np.sum(pix_vecs.T * dipole_dir, axis=1)
        #costheta = np.dot(hp.pix2vec(nside, pix), dipole_dir)
        print("costheta =",costheta)
        print("costheta.shape =",costheta.shape)
        return 1 + A * costheta
    
    for z_min, z_max in zip(z_bins[:-1], z_bins[1:]):
        in_bin = (z > z_min) & (z < z_max)
        
        ipix = hp.ang2pix(nside, np.radians(90-dec[in_bin]), np.radians(ra[in_bin]), nest=nest)
        counts = np.bincount(ipix, weights[in_bin], minlength=npix)
        
        try:
            popt, pcov = curve_fit(model, np.arange(npix), counts, p0=[0.001, 264, 48])
            results.append(popt)
        except:
            results.append([np.nan, np.nan, np.nan])
    return results
````

## Alternative version proposed by Gemini:

````python
def fit_dipole_tomographic_corrected(ra, dec, z, weights, z_bins=[0.1, 0.5, 1.0, 2.0], nest=True):
    results = []
    nside = 32
    npix = hp.nside2npix(nside)
    
    # On définit le modèle à l'extérieur ou on passe nside explicitement
    def model(pix, A, l, b):
        pix_vecs = np.array(hp.pix2vec(nside, pix.astype(int), nest=nest))
        dipole_dir = hp.ang2vec(np.radians(90-b), np.radians(l)) # shape: (3,)
        costheta = np.sum(pix_vecs.T * dipole_dir, axis=1)
        print("costheta =",costheta)
        print("costheta.shape =",costheta.shape)
        return (1 + A * costheta)

    for z_min, z_max in zip(z_bins[:-1], z_bins[1:]):
        in_bin = (z >= z_min) & (z < z_max)
        if np.sum(in_bin) == 0:
            results.append([np.nan, np.nan, np.nan])
            continue
            
        # Création de la carte de comptage (skymap)
        ipix = hp.ang2pix(nside, np.radians(90-dec[in_bin]), np.radians(ra[in_bin]), nest=nest)
        counts = np.bincount(ipix, weights[in_bin], minlength=npix).astype(float)
        
        mean_counts = np.mean(counts)
        pixel_indices = np.arange(npix)

        try:
            # p0: A (amplitude), l (longitude), b (latitude)
            # On ajoute des bornes (bounds) pour aider l'algorithme
            popt, pcov = curve_fit(model, pixel_indices, counts, 
                                   p0=[0.01, 264, 48],
                                   bounds=([0, 0, -90], [1, 360, 90]))
            results.append(popt)
        except Exception as e:
            print(f"Erreur lors du fit pour le bin {z_min}-{z_max}: {e}")
            results.append([np.nan, np.nan, np.nan])
            
    return results
````

# --- Step 6: Visualization ---

````python
def plot_skymap(ra, dec, weights, title):
    nside = 64
    ipix = hp.ang2pix(nside, np.radians(90-dec), np.radians(ra), nest=True)
    m = np.zeros(hp.nside2npix(nside))
    np.add.at(m, ipix, weights)
    hp.mollview(m, title=title, nest=True, cmap='viridis')
    plt.savefig('skymap.png', dpi=200, bbox_inches='tight')
    #plt.close()

def plot_dipole_fits(results, z_bins):
    plt.errorbar(0.5*(np.array(z_bins[:-1])+np.array(z_bins[1:])), [r[0] for r in results], \
                 yerr=[0.0005]*len(results), fmt='o')  # Placeholder error
    plt.xlabel('Redshift')
    plt.ylabel('Dipole Amplitude')
    plt.title('Tomographic Dipole Fit')
    plt.grid(True)
    plt.savefig('dipole_fit.png', dpi=200, bbox_inches='tight')
    #plt.close()
````
